```python
import yfinance as yf
import mplfinance as mpf
```

```python
# Baixando os dados das ações da Apple
dados_apple = yf.download('AAPL', start='2023-01-01', end='2023-12-31')

# Criando o gráfico de candlestick
mpf.plot(dados_apple.head(30), type='candle', figsize=(16, 8), volume=True, mav=(7, 14))
```
